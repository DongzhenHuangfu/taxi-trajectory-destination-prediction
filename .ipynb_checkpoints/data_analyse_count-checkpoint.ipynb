```python
import pandas as pd
import seaborn as sb
import numpy as np
import pickle
import matplotlib.pyplot as plt
import sys
```

```python
with open('./endpoint_per_calltype.pickle', 'rb') as file:
    data_dict = pickle.load(file)
    copy = data_dict['index_label']
```

```python
area = []
with open('./deep_net/data/train_for_deepstructure_area.pickle', 'rb') as file:
    data_dict = pickle.load(file)
    area = data_dict['area']
    
def str2array(string):
    
    ret = []
    
    if(len(string) > 2):  
        together = ''
        for i in string:
            if(i != '['):
                together += i

        first = together.split("],")

        for i in range(len(first)):
            if(i==len(first)-1):
                second = ''
                for j in first[i]:
                    if(j!=']'):
                        second += j
                second = second.split(",")
            else:
                second = first[i].split(",")
            ret.append([float(second[0]), float(second[1])])
        return np.array(ret[-1])
    
    else: return np.array(ret)

def tran_label(string):
    
    point = str2array(string)
    
    if(len(point) == 0):
        return 999999
    else:
        for i in range(len(area) - 1):
            if((point[0] < area[i][0]) & (point[1] < area[i][1])):
                find = True 
                return i   
```

```python
labels = copy["POLYLINE"].apply(tran_label)
```

```python
length = len(labels)
label_dic = {"a": [], "b": [], "c": []}
label_count = {"a": [], "b": [], "c": []}

for i in range(length):
    sys.stdout.write("appending the %dth point of %d, %f%%   \r" % (i, length, ((i+1)/length*100)))
    sys.stdout.flush()
    index = labels[i]
    call_type = copy["CALL_TYPE"][i]
    
    if(call_type == "A"):
        if(index in label_dic["a"]):
            position = label_dic["a"].index(index)
            label_count["a"][position] += 1
        else:
            label_dic["a"].append(index)
            label_count["a"].append(1)
            
    elif(call_type == "B"):
        if(index in label_dic["b"]):
            position = label_dic["b"].index(index)
            label_count["b"][position] += 1
        else:
            label_dic["b"].append(index)
            label_count["b"].append(1)
            
    elif(call_type == "C"):
        if(index in label_dic["c"]):
            position = label_dic["c"].index(index)
            label_count["c"][position] += 1
        else:
            label_dic["c"].append(index)
            label_count["c"].append(1)
            
    else:
        print("index %f has no type!" % index)
```
